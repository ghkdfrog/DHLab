```python
import torchOptics.optics as tt
import warnings 
import torch.nn as nn
import torch
from torchvision.utils import save_image
warnings.filterwarnings('ignore') ## ignore warnings
meta = {'wl' : (638e-9, 515e-9, 450e-9), 'dx':(6.3e-6, 6.3e-6)}
timg = tt.imread('dataset/pokemon.png', meta = meta)
field = timg.clone()

tt.show(field)
```

```python
sim = tt.simulate(field, 20e-3)
tt.show(sim)
```

```python
field = timg.clone()
phase = torch.angle(field)
amp = field.abs()
phase.requires_grad = True
optimizer = torch.optim.Adam([phase], lr=0.001)
criterion = torch.nn.L1Loss()
for i in range(10000):
    optimizer.zero_grad()
    # loss function
    re_field = 1 * torch.exp(1j*phase)
    sim = tt.simulate(re_field, 20e-3)
    loss = criterion(sim, timg.abs())
    loss.backward()
    optimizer.step()
result = tt.simulate(1 * torch.exp(1j*phase), 20e-3)
tt.show(result)
```
